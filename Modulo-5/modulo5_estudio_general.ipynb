## 📋 Referencia Rápida

| Función / Clase | Qué hace |
|---|---|
| `detectar_puertos()` | Escanea puertos serie disponibles |
| `conectar_arduino(puerto, baudrate)` | Abre conexión serial con Arduino |
| `ArduinoSimulado()` | Simula Arduino en software (sin hardware) |
| `tabla_verdad(func, nombre, n_entradas)` | Genera tabla de verdad de una compuerta |

| `circuito_sumador_completo(a, b, c_in)` | Suma 3 bits → suma + acarreo |
| `perceptron_compuerta(tipo)` | Entrena perceptrón para AND/OR/NAND/NOR/NOT |
| `graficar_convergencia(history)` | Curva de errores durante entrenamiento |
| `graficar_frontera_decision(W, b, X, y)` | Frontera de decisión 2D del perceptrón |
| `exportar_perceptron_a_c(W, b, nombre)` | Genera código C++ para Arduino |
| `exportar_modelo_arduino_ino(W, b, nombre)` | Sketch .ino completo |
| `mapear_0_1(valor, min, max)` | Normaliza sensor a rango [0, 1] |
| `binario_a_decimal(bits)` / `decimal_a_binario(n, bits)` | Conversión base |
