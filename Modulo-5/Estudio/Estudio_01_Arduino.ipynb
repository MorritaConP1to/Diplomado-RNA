## Arquitectura Arduino

Arduino es una plataforma de microcontrolador open-source basada en hardware y software simples.

### Componentes clave:
- **Microcontrolador**: ATmega328P (Arduino Uno) — 8-bit, 16 MHz
- **GPIO**: 14 pines digitales (D0–D13), 6 entradas analógicas (A0–A5)
- **Comunicación**: UART (Serial), I²C, SPI
- **Alimentación**: 5V / 3.3V, entrada 7–12V

### Conceptos básicos de C++ para Arduino:
- `setup()` → se ejecuta una vez al iniciar
- `loop()` → se ejecuta en ciclo infinito
- `pinMode(pin, modo)` → configura dirección del pin
- `digitalWrite(pin, HIGH/LOW)` → escribe salida digital
- `digitalRead(pin)` → lee entrada digital
- `analogRead(pin)` → lee entrada analógica (0–1023)
- `analogWrite(pin, valor)` → escribe PWM (0–255)
- `Serial.println()` → envía texto por puerto serie

### Diagrama de pines (Arduino Uno):
```
                   ┌─────────────┐
   (RESET) D0 ─────┤             ├───── D13 (SCK)
   (RX)    D1 ─────┤             ├───── D12 (MISO)
   (TX)    D2 ─────┤   ATmega    ├───── D11 (MOSI)
            D3 ─────┤   328P      ├───── D10 (SS)
            D4 ─────┤             ├───── D9
            D5 ─────┤             ├───── D8
            D6 ─────┤             ├───── D7
            A0 ─────┤             ├───── A5 (SCL)
            A1 ─────┤             ├───── A4 (SDA)
            A2 ─────┤             ├───── A3
                   └─────────────┘
```

## Simulación de Arduino en Python

Usaremos la clase `ArduinoSimulado` para practicar sin necesidad de hardware físico.

````python
arduino = ArduinoSimulado('MiArduino')
arduino.configurar(9600)
````

````python
arduino.pin_mode('LED_BUILTIN', 'OUTPUT')
arduino.pin_mode('D9', 'OUTPUT')
arduino.pin_mode('D2', 'INPUT')
````

````python
arduino.digital_write('LED_BUILTIN', 1)
arduino.digital_write('D9', 1)
arduino.estado()
````

````python
print("=== Blink Pattern (5 ciclos) ===")
for i in range(5):
    print(f"--- Ciclo {i+1} ---")
    arduino.digital_write('LED_BUILTIN', 1)
    arduino.delay(500)
    arduino.estado()
    arduino.digital_write('LED_BUILTIN', 0)
    arduino.delay(500)
    arduino.estado()
print("✅ Blink completado")
````

````python
print("=== Lectura Analógica Simulada ===")
for _ in range(3):
    valor = arduino.analog_read('A0')
    voltaje = valor * 5.0 / 1023.0
    arduino.serial_println(f"A0 = {valor} ({voltaje:.2f}V)")
print("--- Buffer serial ---")
while arduino.serial_available() > 0:
    msg = arduino.serial_read()
    print(f"  Leído: {msg}")
````

## Sketch C++ para Arduino IDE

Copia y pega este código directamente en el Arduino IDE:

```cpp
/*
  Blink Pattern — Ejemplo básico
  Parpadea el LED incorporado cada 500ms
*/

void setup() {
  pinMode(LED_BUILTIN, OUTPUT);
  Serial.begin(9600);
  Serial.println("Arduino listo!");
}

void loop() {
  digitalWrite(LED_BUILTIN, HIGH);
  Serial.println("LED ON");
  delay(500);

  digitalWrite(LED_BUILTIN, LOW);
  Serial.println("LED OFF");
  delay(500);
}
```

### Sketch con lectura analógica:

/*
  Lectura analógica — Lee el potenciómetro en A0
*/

void setup() {
  Serial.begin(9600);
}

void loop() {
  int valor = analogRead(A0);

  Serial.print("Valor: ");
  Serial.print(valor);
  Serial.print("  Voltaje: ");
  Serial.println(voltaje);
  delay(500);
}
