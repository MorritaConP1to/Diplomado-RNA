```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Datos
datos = {
    'Edad': [20, 25, 40, 50],
    'BMI': [22.5, 27.1, 31.0, 29.5],
    'Seguro': [2000, 3500, 8000, 12000]
}

# DataFrame
df = pd.DataFrame(datos)

# Variables independientes
X = df[['Edad', 'BMI']]

# Variable dependiente
y = df['Seguro']

# Crear modelo
modelo = LinearRegression()

# Entrenar modelo
modelo.fit(X, y)

# Predicción
edad_nueva = 30
bmi_nuevo = 28

prediccion = modelo.predict([[edad_nueva, bmi_nuevo]])

print("Costo estimado del seguro:", prediccion[0])
```
