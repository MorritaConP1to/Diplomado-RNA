Importamos la libreria para las funciones matematicas

```python
import numpy as np
```

Esta es la definicion de las clases
En el escalon unitario la salida es 1 cuando la entrada es mayor o igual a 0

```python
def escalon_unitario(x):
    """Esta función devuelve 1 si x es mayor o igual a 0, y 0 en caso contrario."""
    if x >= 0:
        return 1
    else:
        return 0
```

```python
def sigmoide(x):
    """Esta función devuelve un valor entre 0 y 1, dependiendo del valor de x."""
    return 1 / (1 + np.exp(-x))
```

```python
def tangente_hiperbolica(x):
    """Esta función devuelve un valor entre -1 y 1, dependiendo del valor de x."""
    return np.tanh(x)
```

```python
def relu(x):
    """Esta función devuelve x si x es mayor que 0, y 0 en caso contrario."""
    return max(0, x)
```

En este bloque programaremos el proceso de un perceptron
Las 2 tareas importantes son:

    2.


```python
#Cada fila de esta matriz representa una combinacion de entradas y cada columna
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
```

```python
#Los targets son los resultados que esperamos obtener para cada combinacion de entradas
Salida_OR = np.array([[0],[1],[1],[1]])
Salida_AND = np.array([[0],[0],[0],[1]])
Salida_NOR = np.array([[1],[0],[0],[0]])
Salida_NAND = np.array([[1],[1],[1],[0]])
Salida_NOT = np.array([[1],[0],[0],[0]])
Salida_XOR = np.array([[0],[1],[1],[0]])
#targets = np.hstack((Salida_OR, Salida_AND, Salida_NOR, Salida_NAND, Salida_NOT, Salida_XOR))
targets = Salida_OR
```

En esta variable de tipo lista almacenaremos la salidas del perceptron

```python
outputs = []
```

Modificamos el valor de los pesos (weights) y del bias del perceptron para que al hacer la suma ponderada (o combinacion lineal) y llama a la funcion de activacion del perfecptron sea igual a los targets

```python
weights = np.array([[1],[-1]])
bias = -0.5
```

Para cada entrada ejecuta el procedimiento del perceptron y almacena la salida en la variable outputs

```python
for i in range(targets.shape[0]):
    #Calcula la suma ponderada (o combinacion lineal) de las entradas y los pesos, y luego agrega el sesgo
    h = np.dot(inputs[i], weights) + bias
    #Aplica la funcion de activacion escalon unitario a la suma ponderada para
    y = escalon_unitario(h) # Aqui puedes cambiar la funcion de activacion a sigmoide(h), tangente_hiperbolica(h) o relu(h)
    outputs.append([h,y])
    
```

```python
# Imprime las salidas del perceptron para cada combinacion de entradas
print(' Input 1',' Input 2',' Combinacion lineal',' target',' Output')
for i in range(len(outputs)):
    print(inputs[i][0],'       ', inputs[i][1],'               ', outputs[i][0][0],'       ', targets[i][0],'    ', outputs[i][1])
    
```
