## 🔄 Pipeline completo de Machine Learning

```
 1. Cargar datos          → obtener X (características) e Y (etiquetas)
 2. Explorar (EDA)        → entender qué hay antes de modelar
 3. Seleccionar features  → quedarse con las columnas útiles
 4. Dividir               → train / test (nunca mezclarlos)
 5. Escalar               → poner todos los valores en la misma escala
 6. Crear modelo          → ← aquí cambia según el tema de la clase
 7. Entrenar              → modelo.fit(X_train, y_train)
 8. Predecir              → Y_pred = modelo.predict(X_test)
 9. Evaluar               → medir accuracy, ver errores
10. Comparar              → comparar varios modelos entre sí
```

Los pasos 1-5 y 7-10 son casi idénticos siempre.  
**Solo cambia el paso 6** según qué tema estés viendo.

## 🤖 4. Perceptrón (sklearn) — Clasificación binaria

## 📈 5. Regresión Logística — Clasificación binaria

````python
# MACHOTE — Regresión Logística con breast_cancer

X_cancer, y_cancer = cargar_dataset_sklearn('breast_cancer')

X_tr_c, X_te_c, y_tr_c, y_te_c = train_test_split(X_cancer, y_cancer, test_size=0.25, random_state=42, stratify=y_cancer)
X_tr_c_sc, X_te_c_sc, _ = escalar_datos(X_tr_c, X_te_c, metodo='standard')

modelo_log = crear_regresion_logistica(C=1.0, max_iter=1000, random_state=42)
modelo_log.fit(X_tr_c_sc, y_tr_c)

y_pred_log = modelo_log.predict(X_te_c_sc)
acc_log = accuracy_score(y_te_c, y_pred_log)

print(f'Regresión Logística — Accuracy: {acc_log:.4f} ({acc_log*100:.1f}%)')
print()

# Probabilidades (lo especial de logística)
probas = modelo_log.predict_proba(X_te_c_sc)
print('Primeras 5 predicciones con probabilidades:')
for i in range(5):
    print(f'  Pred={y_pred_log[i]} Real={y_te_c[i]} | P(0)={probas[i][0]:.3f} P(1)={probas[i][1]:.3f}')

evaluar_clasificacion(y_te_c, y_pred_log, clases=['Maligno', 'Benigno'])
````

## ⚡ 7. SVM — Comparación de kernels

````python
# MACHOTE — SVM: comparar kernels linear, rbf y poly

kernels = ['linear', 'rbf', 'poly']
resultados_svm = {}

for kernel in kernels:
    modelo_svm = crear_svm(kernel=kernel, C=1.0, random_state=42)
    modelo_svm.fit(X_tr_c_sc, y_tr_c)
    y_pred = modelo_svm.predict(X_te_c_sc)
    acc = accuracy_score(y_te_c, y_pred)
    resultados_svm[kernel] = acc
    print(f'SVM ({kernel}) — Accuracy: {acc:.4f} ({acc*100:.1f}%)')

# Gráfica comparativa
plt.figure(figsize=(6, 4))
plt.bar(resultados_svm.keys(), [v*100 for v in resultados_svm.values()], color=['#3B8BD4', '#D85A30', '#4CAF50'])
plt.ylabel('Accuracy (%)')
plt.title('Comparación de kernels SVM')
for i, (k, v) in enumerate(resultados_svm.items()):
    plt.text(i, v*100 + 0.5, f'{v*100:.1f}%', ha='center', fontweight='bold')
plt.tight_layout()
plt.show()

print(f'\nMejor kernel: {max(resultados_svm, key=resultados_svm.get)}')
````

## 🧠 12. PerceptronManual — Demostración de XOR

````python
# MACHOTE — PerceptronManual: entrenar con compuerta OR

X_or = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_or = np.array([0, 1, 1, 1])  # OR

perc_or = PerceptronManual(tasa_aprendizaje=0.1, epochs=20)
perc_or.entrenar(X_or, y_or)

pred_or = perc_or.predecir(X_or)
print(f'OR - Predicciones: {pred_or}')
print(f'OR - Esperado:     {y_or}')
print(f'OR - Accuracy:     {accuracy_score(y_or, pred_or)*100:.1f}%')
````

````python
# MACHOTE — XOR: problema linealmente no separable
# El perceptrón simple NO puede resolver XOR (necesita capas ocultas)

X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])  # XOR

perc_xor = PerceptronManual(tasa_aprendizaje=0.1, epochs=50)
perc_xor.entrenar(X_xor, y_xor)

pred_xor = perc_xor.predecir(X_xor)

print(f'XOR - Predicciones: {pred_xor}')
print(f'XOR - Esperado:     {y_xor}')
print(f'XOR - Accuracy:     {accuracy_score(y_xor, pred_xor)*100:.1f}%')
print()
print('⚠️ El perceptrón simple falla en XOR porque las clases no son')
print('   linealmente separables. Se necesitan capas ocultas (MLP).')
````

## 🔍 Solución de Errores Comunes

| Error | Causa probable | Solución |
|---|---|---|
| `ImportError: No module named 'libreria_modulo3'` | Ruta incorrecta | Verificar `sys.path.insert(0, ...)` apunte al padre |
| `ValueError: Shapes incompatible` | Dimensiones incorrectas | Verificar `X.shape` y `y.shape` |
| Accuracy muy bajo en SVM | Datos sin escalar | Llamar `escalar_datos()` antes de entrenar |
| Overfitting (train perfecto, test malo) | Árbol muy profundo | Reducir `max_depth` o aumentar `min_samples_split` |
| K-Means da clusters pobres | K incorrecto | Usar `encontrar_k_optimo()` para elegir K |
| Perceptrón no converge | Datos no separables linealmente | Usar otro modelo (SVM, Árbol, MLP) |
| NaN en regresión | Datos con NaN sin limpiar | Revisar valores nulos con `df.isnull().sum()` |
| `FileNotFoundError` al cargar CSV | Ruta incorrecta | Usar ruta relativa o `os.path.join()` |

> 💡 **Regla de oro:** Si el notebook falla, lee el **último error primero**.  
> Googlea el mensaje exacto — probablemente alguien ya lo resolvió en Stack Overflow.
