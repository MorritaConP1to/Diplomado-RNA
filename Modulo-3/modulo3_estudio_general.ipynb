## 🧠 Tema 1: Perceptrón

El **Perceptrón** es la unidad fundamental de las redes neuronales: recibe entradas, las multiplica por pesos, suma un sesgo y aplica una función escalón. scikit-learn implementa `Perceptron` como clasificador lineal binario.

```python
# Perceptrón con Iris (2 clases: setosa vs versicolor)
X, y = cargar_dataset_sklearn('iris')
X_bin, y_bin = X[:100], y[:100]
X_train, X_test, y_train, y_test = dividir_datos(X_bin, y_bin, test_size=0.3)

modelo = crear_perceptron(max_iter=1000)
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

## 📈 Tema 2: Regresión Logística

La **Regresión Logística** extiende el perceptrón usando la función sigmoide para producir probabilidades entre 0 y 1. Aunque se llama "regresión", es un método de clasificación binaria y multiclase.

```python
# Regresión Logística con breast_cancer
X, y = cargar_dataset_sklearn('breast_cancer')
X_train, X_test, y_train, y_test = dividir_datos(X, y, test_size=0.2)

modelo = crear_regresion_logistica(max_iter=3000)
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)

evaluar_clasificacion(y_test, y_pred, clases=['Maligno', 'Benigno'])
```

## 🌳 Tema 3: Árbol de Decisión

Un **Árbol de Decisión** particiona el espacio con preguntas binarias (si/no). Cada hoja asigna una clase. Es fácil de interpretar, no requiere escalamiento, y se puede visualizar directamente.

```python
# Árbol de Decisión con Iris completo
X, y = cargar_dataset_sklearn('iris')
X_train, X_test, y_train, y_test = dividir_datos(X, y, test_size=0.2)

modelo = crear_arbol_decision(max_depth=3)
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")

visualizar_arbol(modelo,
    feature_names=['sepal_l', 'sepal_w', 'petal_l', 'petal_w'],
    class_names=['setosa', 'versicolor', 'virginica'])
```

## 🔷 Tema 4: SVM (Máquinas de Soporte Vectorial)

**SVM** busca el hiperplano que maximiza el margen entre clases. Con kernels (lineal, RBF, polinomial) resuelve problemas no lineales transformando el espacio de características.

```python
# Comparación rápida de kernels en Iris
X, y = cargar_dataset_sklearn('iris')
X_train, X_test, y_train, y_test = dividir_datos(X, y, test_size=0.2)

for kernel in ['linear', 'rbf', 'poly']:
    modelo = crear_svm(kernel=kernel)
    modelo.fit(X_train, y_train)
    acc = accuracy_score(y_test, modelo.predict(X_test))
    print(f"kernel={kernel:8s} → Accuracy: {acc:.4f}")
```

## 📉 Tema 5: Regresión Lineal

La **Regresión Lineal** modela la relación entre una variable dependiente y una o más independientes ajustando un hiperplano que minimiza el error cuadrático medio (MSE).

```python
# Regresión Lineal con diabetes
X, y = cargar_dataset_sklearn('diabetes')
X_train, X_test, y_train, y_test = dividir_datos(X, y, test_size=0.2)

modelo = crear_regresion_lineal()
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)

evaluar_regresion(y_test, y_pred)
graficar_predicciones(y_test, y_pred, 'Regresión Lineal — Diabetes')
```

## 🔵 Tema 6: K-Means (Clustering)

**K-Means** agrupa datos en `k` clusters minimizando la distancia intra-cluster. El método del codo grafica la inercia vs k para elegir el número óptimo de clusters.

```python
# K-Means: método del codo + clustering en Iris
X, y = cargar_dataset_sklearn('iris')
encontrar_k_optimo(X, k_max=8)

kmeans = crear_kmeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(X)

print(f"Silhouette Score: {silhouette_score(X, y_kmeans):.4f}")
print(f"Centroides:\n{kmeans.cluster_centers_}")
```
