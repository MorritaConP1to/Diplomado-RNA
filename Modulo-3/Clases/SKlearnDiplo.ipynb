```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# Datos de emtrada a modelo 
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
targets = np.array([0,1,1,1])

# Modelo
modelo = Perceptron(eta0=0.1, max_iter=1, warm_start=True) #eta0 indie de aprendizaje #max:iter numero maximo de iteracion con los daros de entrada  #warm memoria de pesos por cada epoca


epochs = 15
errores = []

#Aprendizaje - entrenamiento
modelo.partial_fit(inputs, targets, classes=np.array([0,1])) #entrenamiento por epoca  #clases diferententes clases que esperamos 

print("\nPesos Iniciales")
print("Weights:", modelo.coef_)
print("Bias:", modelo.intercept_)

#aprendizaje
for epoch in range(epochs):
    modelo.partial_fit(inputs, targets)

    predicciones = modelo.predict(inputs)
    error = np.sum(predicciones != targets)
    errores.append(error)

 # ver epocas
    print(f"\nÉpoca {epoch+1}")
    print("Predicciones:", predicciones)
    print("Error:", error)
    print("Weights:", modelo.coef_)
    print("Bias:", modelo.intercept_)

#Finales
print("\nPesos Finales")
print("Weights:", modelo.coef_)
print("Bias:", modelo.intercept_)

plt.figure()
plt.plot(range(1, epochs+1), errores, marker='o')
plt.xlabel("Épocas")
plt.ylabel("Número de errores")
plt.title("Error del Perceptrón (Scikit-learn)")
plt.grid()
plt.show()
```
