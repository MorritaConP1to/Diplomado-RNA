```python
import numpy as np

def escalon(x):
    return 1 if x >= 0 else 0

# Perceptrón OR
def OR(x):
    w = np.array([1,1])
    b = -0.5
    return escalon(np.dot(x, w) + b)

# Perceptrón NAND
def NAND(x):
    w = np.array([-1,-1])
    b = 1.5
    return escalon(np.dot(x, w) + b)

# Perceptrón AND
def AND(x):
    w = np.array([1,1])
    b = -1.5
    return escalon(np.dot(x, w) + b)

# XOR usando los anteriores
def XOR(x):
    salida_or = OR(x)
    salida_nand = NAND(x)
    return AND(np.array([salida_or, salida_nand]))

# Prueba
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])

print("x1 x2 -> XOR")
for x in inputs:
    print(x[0], x[1], "->", XOR(x))
```
