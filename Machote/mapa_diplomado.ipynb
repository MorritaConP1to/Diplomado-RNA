### Métricas importantes del Tema 3
```
Para CLASIFICACIÓN:
  Accuracy  = (TP + TN) / total          ← % de aciertos
  Precision = TP / (TP + FP)             ← de los que predije positivo, ¿cuántos lo eran?
  Recall    = TP / (TP + FN)             ← de todos los positivos reales, ¿cuántos encontré?
  F1        = 2 * (Precision*Recall) / (Precision+Recall)  ← balance entre ambas

Para REGRESIÓN:
  MSE  = promedio de (real - predicho)²  ← error cuadrático medio
  RMSE = √MSE                            ← mismas unidades que Y
  R²   = 1 - (MSE / varianza_Y)         ← 1.0=perfecto, 0.0=no explica nada
```

### Gradiente Descendente
```
¿Cómo aprende la regresión?

1. Inicia con pesos aleatorios (o en 0)
2. Calcula el error con esos pesos
3. Calcula el gradiente (dirección donde el error SUBE más)
4. Mueve los pesos en la dirección CONTRARIA (hacia donde baja)
5. Repite hasta que el error converge (deja de bajar)

Es como bajar una colina caminando siempre hacia donde el suelo baja más rápido.
La tasa de aprendizaje (lr) controla qué tan grandes son los pasos.

w_nuevo = w_viejo - lr * gradiente
```

## Tema 1 — Fundamentos de Redes Neuronales

### Evolución: MPNeuron → Perceptrón → MLP
```
MPNeuron (1943):
  - Entradas binarias
  - Sin pesos, sin aprendizaje
  - Threshold por fuerza bruta

Perceptrón (1957):
  - Entradas continuas
  - Pesos que se aprenden
  - Una sola capa
  - Solo puede separar clases linealmente separables

MLP - Perceptrón Multicapa (1986+):
  - Varias capas ocultas
  - Funciones de activación no lineales
  - Puede aprender patrones complejos (XOR, imágenes, texto)
  - Aprende con backpropagation + gradient descent
```

### Funciones de Activación
Ver: `Machote_ML/Funciones_Activacion.ipynb`

```python
# Las funciones de activación más importantes:

# Escalón (Step) — usada en MPNeuron y Perceptrón
# salida = 1 si z >= 0, sino 0
# Problema: no es diferenciable → no funciona con gradient descent

# Sigmoide — usada en regresión logística y capas de salida binarias
# salida = 1 / (1 + e^(-z))  → rango (0, 1)
# Problema: gradientes muy pequeños en los extremos (vanishing gradient)

# ReLU — la más usada en capas ocultas de redes modernas
# salida = max(0, z)          → rango (0, ∞)
# Ventaja: simple, eficiente, resuelve vanishing gradient

# Tanh — similar a sigmoide pero centrada en 0
# salida = (e^z - e^(-z)) / (e^z + e^(-z))  → rango (-1, 1)

# Softmax — para clasificación multiclase (capa de salida)
# Convierte un vector de números a probabilidades que suman 1
```

### MLP con Keras/TensorFlow
```python
from tensorflow import keras

# Construir el modelo capa por capa
modelo = keras.Sequential([
    # Capa de entrada: especifica cuántas features tiene X
    keras.layers.Input(shape=(n_features,)),

    # Capas ocultas: Dense = cada neurona conectada a todas las anteriores
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),

    # Capa de salida
    keras.layers.Dense(1, activation='sigmoid')   # binaria
    # keras.layers.Dense(3, activation='softmax')  # multiclase (3 clases)
])

# Configurar cómo va a aprender
modelo.compile(
    optimizer='adam',              # algoritmo de gradient descent
    loss='binary_crossentropy',    # función de pérdida para clasificación binaria
    metrics=['accuracy']
)

# Entrenar
historial = modelo.fit(
    X_train_sc, y_train,
    epochs=50,              # cuántas veces pasa por todo el dataset
    batch_size=32,          # cuántas muestras procesa a la vez
    validation_split=0.2,   # 20% del train para validar
    verbose=1
)

# Graficar curvas de aprendizaje
plt.plot(historial.history['accuracy'],     label='train')
plt.plot(historial.history['val_accuracy'], label='validación')
plt.legend()
plt.show()
```

## Tema 5 — Entrenamiento y conceptos clave
```
Overfitting (sobreajuste):
  El modelo memoriza el train pero falla en datos nuevos.
  train accuracy alta, val accuracy baja.
  Soluciones: más datos, Dropout, regularización, reducir capas.

Underfitting (subajuste):
  El modelo no aprende suficiente.
  train accuracy baja y val accuracy baja.
  Soluciones: más épocas, más capas, más neuronas.

Hiperparámetros (tú los decides, el modelo no los aprende):
  - Tasa de aprendizaje (lr)
  - Número de capas y neuronas
  - Épocas y batch size
  - Función de activación
  - Regularización (L1, L2, Dropout)

Parámetros (el modelo los aprende):
  - Pesos (weights)
  - Bias
```

```cpp
// Código en C++ para Arduino (IDE de Arduino, no Python)
void setup() {
    pinMode(13, OUTPUT);   // configura pin 13 como salida
    Serial.begin(9600);    // comunicación serial a 9600 baudios
}

void loop() {
    if (Serial.available()) {      // si llegan datos de Python
        char dato = Serial.read(); // leer el dato
        if (dato == '1') {
            digitalWrite(13, HIGH); // encender LED
        } else {
            digitalWrite(13, LOW);  // apagar LED
        }
    }
}
```
