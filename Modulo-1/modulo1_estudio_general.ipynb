## Vectores

Un **vector** es un arreglo ordenado de números que representa magnitud y dirección. Operaciones clave:
- **Suma / Resta**: elemento a elemento
- **Producto punto** (v · w): Σ v_i w_i — mide similitud entre vectores
- **Magnitud** (norma L2): ‖v‖ = √(Σ v_i²)
- **Ángulo**: cos θ = (v · w) / (‖v‖ ‖w‖)

## Matrices

Una **matriz** es un arreglo bidimensional de números. En RNA las matrices representan los **pesos** de las conexiones entre capas. Operaciones fundamentales:
- **Suma / Resta**: elemento a elemento (mismas dimensiones)
- **Producto matricial** (A_m×n · B_n×p = C_m×p)
- **Transpuesta** (A^T): intercambia filas por columnas
- **Determinante** e **Inversa** (matrices cuadradas)

## Derivadas

Las derivadas son esenciales para el **gradiente descendente**, algoritmo que optimiza los pesos de una red neuronal. El gradiente indica la dirección de mayor cambio del error; actualizamos los pesos en dirección opuesta.

| Función | Derivada |
|---------|----------|
| x^n | n·x^(n-1) |
| e^x | e^x |
| ln(x) | 1/x |
| σ(x) (sigmoide) | σ(x)·(1 − σ(x)) |
| tanh(x) | 1 − tanh²(x) |

La **regla de la cadena** es la base del backpropagation: (f ∘ g)' = f'(g(x)) · g'(x).

## Glosario Rápido

| Término | Definición |
|---------|-----------|
| **IA** | Campo que busca crear sistemas con inteligencia similar a la humana |
| **Machine Learning** | Subcampo de IA donde los modelos aprenden patrones a partir de datos |
| **Deep Learning** | Subcampo de ML que usa redes neuronales profundas (múltiples capas) |
| **Red Neuronal** | Modelo inspirado en el cerebro, compuesto de neuronas artificiales en capas |
| **Perceptrón** | Unidad básica de una red neuronal: entradas → pesos → suma → activación |
| **Backpropagation** | Algoritmo que ajusta pesos propagando el error hacia atrás por la red |
| **Gradiente Descendente** | Optimización que minimiza el error ajustando pesos en dirección opuesta al gradiente |
| **Overfitting** | El modelo memoriza los datos de entrenamiento pero falla en datos nuevos |
| **CNN** | Red convolucional especializada en datos con estructura de cuadrícula (imágenes) |
| **RNN / LSTM** | Red recurrente diseñada para procesar datos secuenciales (texto, series temporales) |
| **Transformer** | Arquitectura basada en atención que revolucionó el NLP (base de GPT, BERT) |
