## 1. ¿Qué es IA, ML, DL?

### Diagrama de Venn

```
┌─────────────────────────────────────────────┐
│               INTELIGENCIA ARTIFICIAL        │
│  ┌────────────────────────────────────┐     │
│  │        MACHINE LEARNING            │     │
│  │  ┌───────────────────────────┐     │     │
│  │  │    DEEP LEARNING          │     │     │
│  │  │  ┌─────────────────┐      │     │     │
│  │  │  │  Redes          │      │     │     │
│  │  │  │  Neuronales     │      │     │     │
│  │  │  │  Artificiales   │      │     │     │
│  │  │  └─────────────────┘      │     │     │
│  │  └───────────────────────────┘     │     │
│  └────────────────────────────────────┘     │
└─────────────────────────────────────────────┘
```

| Concepto | Definición | Ejemplo |
|---|---|---|
| **IA** (Inteligencia Artificial) | Campo que busca crear sistemas capaces de realizar tareas que requieren inteligencia humana | Ajedrez, asistentes virtuales |
| **ML** (Machine Learning) | Subcampo de IA donde los modelos aprenden patrones a partir de datos sin ser programados explícitamente | Clasificación de spam |
| **DL** (Deep Learning) | Subcampo de ML que usa redes neuronales profundas (múltiples capas) para aprender representaciones jerárquicas | Reconocimiento facial, ChatGPT |
| **RNA** (Red Neuronal Artificial) | Unidad base del DL, inspirada en el cerebro biológico | Perceptrón, CNN, RNN |

> **Regla mnemotécnica:** Todo DL es ML, todo ML es IA, pero NO al revés.

## 2. Tipos de Aprendizaje

| Tipo | Definición | Etiquetas | Retroalimentación | Ejemplos |
|---|---|---|---|---|
| **Supervisado** | El modelo aprende de pares (entrada, etiqueta) | Sí | Error directo (comparación con etiqueta) | Regresión lineal, SVM, Random Forest |
| **No Supervisado** | El modelo encuentra patrones sin etiquetas | No | Similitud interna de los datos | K-Means, PCA, DBSCAN |
| **Semisupervisado** | Combina pocos datos etiquetados + muchos no etiquetados | Pocas | Pseudoetiquetas + error supervisado | Self-training, MixMatch |
| **Refuerzo** | El agente aprende por prueba y error con recompensas | No (usa recompensa) | Señal de recompensa (delayada) | Q-Learning, Deep Q-Networks, AlphaGo |
| **Auto-supervisado** | Genera etiquetas automáticamente desde los datos | Auto-generadas | Contraste / predictivo | BERT, GPT, SimCLR |

```
Supervisado:  (x, y) → aprender f: x → y
No Supervisado:  x → encontrar estructura oculta
Refuerzo:  estado → acción → recompensa → política óptima
```

## 3. Historia de las RNA

| Año | Hito | Autor(es) | Relevancia |
|---|---|---|---|
| 1943 | Primer modelo matemático de neurona (McCulloch-Pitts) | McCulloch & Pitts | Nace el concepto de neurona artificial |
| 1958 | Perceptrón (primer clasificador lineal) | Rosenblatt | Primer algoritmo de aprendizaje supervisado |
| 1969 | *Perceptrons* (libro que prueba limitaciones) | Minsky & Papert | Primer **invierno de la IA** |
| 1974 | Backpropagation propuesto por primera vez | Werbos | Base del entrenamiento multicapa |
| 1986 | Backpropagation popularizado | Rumelhart, Hinton, Williams | Renacimiento de las RNA |
| 1989 | Demostración: RNA universal de 1 capa oculta | Cybenko | Teorema de aproximación universal |
| 1998 | LeNet-5 (CNN para dígitos) | LeCun et al. | Primera CNN práctica |
| 2006 | Deep Belief Networks (pre-entrenamiento no supervisado) | Hinton | Nace el **Deep Learning** moderno |
| 2012 | AlexNet gana ImageNet (GPU + ReLU + Dropout) | Krizhevsky, Sutskever, Hinton | Revolución del DL en visión |
| 2014 | GANs (Redes Generativas Antagónicas) | Goodfellow | Generación de datos sintéticos |
| 2017 | Transformer: *"Attention is All You Need"* | Vaswani et al. | Arquitectura dominante actual |
| 2018 | BERT (pre-entrenamiento bidireccional) | Devlin et al. | SOTA en NLP |
| 2020 | GPT-3 (175B parámetros) | OpenAI | Modelos de lenguaje a gran escala |
| 2022 | ChatGPT / Stable Diffusion | OpenAI / Rombach et al. | IA generativa democratizada |
| 2024 | GPT-4o, Gemini, Claude 3, Sora | Múltiples | Modelos multimodales de frontera |

> **Inviernos de la IA:** 1969-1980 y 1987-1993. Periodos de reducción de financiamiento por expectativas no cumplidas.

## 4. Arquitectura de una Neurona

### Diagrama

```
      x₁ ─── w₁ ──┐
                  │
      x₂ ─── w₂ ──┼──→ Σ ──→ f(·) ──→ y
                  │       ↑          │
      x₃ ─── w₃ ──┘    sesgo (b)     │
                                      │
      ...                              │
                                      │
      xₙ ─── wₙ ──┘                  └── realimentación (entrenamiento)
```

### Componentes

| Componente | Símbolo | Descripción |
|---|---|---|
| Entradas | $x_1, x_2, ..., x_n$ | Señales de entrada (características o activaciones de capa anterior) |
| Pesos | $w_1, w_2, ..., w_n$ | Importancia de cada entrada (aprendida durante entrenamiento) |
| Sesgo | $b$ | Término independiente que desplaza la activación |
| Suma ponderada | $z = \sum_{i=1}^{n} w_i x_i + b$ | Combinación lineal de entradas y pesos |
| Función de activación | $f(z)$ | Transformación no lineal que determina la salida |
| Salida | $y = f(z)$ | Resultado final de la neurona |

### Funciones de Activación Comunes

| Función | Fórmula | Rango | Uso típico |
|---|---|---|---|
| Sigmoide | $\sigma(z) = \frac{1}{1+e^{-z}}$ | (0, 1) | Clasificación binaria (capa de salida) |
| Tanh | $\tanh(z) = \frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$ | (-1, 1) | Capas ocultas (RNN clásicas) |
| ReLU | $\text{ReLU}(z) = \max(0, z)$ | [0, ∞) | Capas ocultas (default moderno) |
| Leaky ReLU | $\text{LReLU}(z) = \max(0.01z, z)$ | (-∞, ∞) | Evita neuronas muertas |
| Softmax | $\sigma(\mathbf{z})_i = \frac{e^{z_i}}{\sum_j e^{z_j}}$ | (0, 1) suma=1 | Clasificación multiclase |

### Cómputo en Forma Vectorial

```
z = wᵀ x + b    (producto punto + sesgo)
y = f(z)         (activación)
```

## 6. Repaso Matemático

### 6.1 Vectores

| Concepto | Fórmula | Notas |
|---|---|---|
| Vector | $\mathbf{v} = [v_1, v_2, ..., v_n]$ | Columna por defecto |
| Producto Punto (Dot) | $\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^{n} a_i b_i$ | Escalar; mide similitud |
| Norma L2 (Euclideana) | $\|\mathbf{v}\|_2 = \sqrt{\sum_i v_i^2}$ | Magnitud del vector |
| Norma L1 (Manhattan) | $\|\mathbf{v}\|_1 = \sum_i |v_i|$ | Robusta a outliers |
| Distancia Euclideana | $d(\mathbf{a},\mathbf{b}) = \|\mathbf{a} - \mathbf{b}\|_2$ | Métrica común |
| Similitud Coseno | $\cos(\theta) = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\|\|\mathbf{b}\|}$ | [-1, 1]; dirección |

### 6.2 Matrices

| Operación | Fórmula | Dimensiones |
|---|---|---|
| Producto | $C_{ij} = \sum_k A_{ik} B_{kj}$ | $(m \times n)(n \times p) \to (m \times p)$ |
| Transpuesta | $(A^T)_{ij} = A_{ji}$ | $(m \times n) \to (n \times m)$ |
| Determinante (2×2) | $\det\begin{pmatrix}a&b\\c&d\end{pmatrix} = ad - bc$ | Solo matrices cuadradas |
| Determinante (3×3) | $\det(A) = a(ei-fh) - b(di-fg) + c(dh-eg)$ | Regla de Sarrus |
| Inversa | $A^{-1}A = AA^{-1} = I$ | Existe solo si $\det(A) \neq 0$ |
| Traza | $\text{tr}(A) = \sum_i A_{ii}$ | Suma de diagonal |
| Identidad | $I_{ij} = \begin{cases}1 & i=j \\ 0 & i \neq j\end{cases}$ | Neutro multiplicativo |

### 6.3 Derivadas

| Regla | Fórmula | Descripción |
|---|---|---|
| Constante | $\frac{d}{dx}c = 0$ | - |
| Potencia | $\frac{d}{dx}x^n = nx^{n-1}$ | - |
| Exponencial | $\frac{d}{dx}e^x = e^x$ | - |
| Logaritmo | $\frac{d}{dx}\ln(x) = \frac{1}{x}$ | - |
| Suma | $\frac{d}{dx}[f+g] = f' + g'$ | Linealidad |
| Producto | $\frac{d}{dx}[f \cdot g] = f'g + fg'$ | - |
| Cociente | $\frac{d}{dx}[f/g] = \frac{f'g - fg'}{g^2}$ | - |
| **Cadena** | $\frac{d}{dx}f(g(x)) = f'(g(x)) \cdot g'(x)$ | **Esencial para backpropagation** |

### 6.4 Derivadas Importantes en DL

| Función | Derivada |
|---|---|
| Sigmoide $\sigma(x) = \frac{1}{1+e^{-x}}$ | $\sigma'(x) = \sigma(x)(1-\sigma(x))$ |
| Tanh | $\tanh'(x) = 1 - \tanh^2(x)$ |
| ReLU | $\text{ReLU}'(x) = \begin{cases}1 & x > 0 \\ 0 & x \leq 0\end{cases}$ |
| MSE (error) | $\frac{\partial}{\partial y}(y-\hat{y})^2 = 2(y-\hat{y})$ |

### 6.5 Regla de la Cadena en Backpropagation

Para una red de 3 capas: $L = f_3(f_2(f_1(x)))$

$$
\frac{\partial L}{\partial w_1} = \frac{\partial L}{\partial f_3} \cdot \frac{\partial f_3}{\partial f_2} \cdot \frac{\partial f_2}{\partial f_1} \cdot \frac{\partial f_1}{\partial w_1}
$$

## 9. Glosario

| Término | Definición |
|---|---|
| **IA** (Inteligencia Artificial) | Campo de la computación que crea sistemas con capacidad de razonamiento similar al humano |
| **ML** (Machine Learning) | Subcampo de IA donde los algoritmos aprenden patrones a partir de datos |
| **DL** (Deep Learning) | Subcampo de ML basado en redes neuronales profundas (múltiples capas ocultas) |
| **RNA** (Red Neuronal Artificial) | Modelo computacional inspirado en neuronas biológicas |
| **Perceptrón** | Unidad básica de una RNA (neurona con pesos, sesgo y activación) |
| **Peso** (Weight) | Parámetro que escala cada entrada; se ajusta durante el entrenamiento |
| **Sesgo** (Bias) | Parámetro que desplaza la activación (término independiente) |
| **Función de Activación** | Transformación no lineal aplicada a la suma ponderada |
| **Forward Propagation** | Pase hacia adelante: cómputo de la salida desde las entradas |
| **Backpropagation** | Algoritmo que propaga el error hacia atrás usando la regla de la cadena |
| **Gradiente** | Vector de derivadas parciales; dirección de mayor ascenso |
| **Gradient Descent** | Algoritmo de optimización que minimiza la función de pérdida |
| **Learning Rate** | Tamaño del paso en cada iteración de descenso de gradiente |
| **Época** (Epoch) | Una pasada completa por todo el conjunto de entrenamiento |
| **Batch** | Subconjunto de datos usado en una iteración de entrenamiento |
| **Overfitting** | El modelo aprende demasiado bien los datos de entrenamiento pero falla en nuevos datos |
| **Underfitting** | El modelo no aprende suficientemente los patrones de los datos |
| **Regularización** | Técnica para reducir overfitting (L1, L2, Dropout, Early Stopping) |
| **Validación Cruzada** (Cross-Validation) | Técnica para evaluar el modelo dividiendo los datos en múltiples pliegues |
| **Normalización** | Escalar datos a un rango estándar (ej. [0,1] o media=0, std=1) |
| **Función de Pérdida** (Loss) | Métrica que cuantifica el error entre predicción y valor real |
| **MSE** (Mean Squared Error) | Error cuadrático medio; pérdida común para regresión |
| **Entropía Cruzada** (Cross-Entropy) | Pérdida común para clasificación |
| **Hiperparámetro** | Parámetro configurado antes del entrenamiento (learning rate, batch size, etc.) |
| **GPU** (Graphics Processing Unit) | Hardware que acelera el entrenamiento de DL (cómputo paralelo) |
