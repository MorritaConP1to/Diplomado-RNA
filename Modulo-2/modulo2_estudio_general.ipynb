## 📦 Tema 3: Estructuras de Datos

```python

coord = (10, 20)

# List comprehension
[x**2 for x in range(5) if x > 2]  # → [9, 16]
```

## 🧠 Tema 5: MP Neuron

### Flujo de trabajo
1. Cargar datos (ej. cáncer de mama)
2. Binarizar: `X_bin = binarizar_datos(X)`
3. Entrenar: `mp.entrenar(X_bin, y)`
4. Predecir: `preds = mp.predecir(X_test)`
5. Evaluar: `accuracy_score(y_test, preds)`

```python
from libreria_modulo2 import MPNeuron, binarizar_datos

mp = MPNeuron()
mp.entrenar(X_train_bin, y_train)
print(f'Precisión: {mp.predecir(X_test_bin)}')
```

````python
# Demo rápida MP Neuron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

np.random.seed(42)
X = np.random.rand(200, 4)
y = (X[:, 0] + X[:, 1] > 1).astype(int)

X_bin = binarizar_datos(X)
Xt, Xv, yt, yv = train_test_split(X_bin, y, test_size=0.2, random_state=42)

mp = MPNeuron()
mp.entrenar(Xt, yt)
pred = mp.predecir(Xv)
print(f'Precisión MP Neuron: {accuracy_score(yv, pred):.2%}')
````
