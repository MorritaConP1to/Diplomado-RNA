```python
# MACHOTE — Boolean Trick (evitar if/else)
# Calcular calorías sin usar if
sexo = 'hombre'
peso = 70

es_hombre = (sexo == 'hombre')  # True = 1, False = 0
calorias = es_hombre * (66 + 13.7 * peso) + (1 - es_hombre) * (655 + 9.6 * peso)
print(f'Calorías: {calorias:.1f}')
```

## 🧠 7. MP Neuron (McCulloch-Pitts)

```python
# MACHOTE — MP Neuron con datos sintéticos
from libreria_modulo2 import MPNeuron, binarizar_datos

np.random.seed(42)
X = np.random.rand(100, 5)
y = (X[:, 0] + X[:, 1] > 1).astype(int)

X_bin = binarizar_datos(X)
X_train, X_test, y_train, y_test = train_test_split(X_bin, y, test_size=0.2, random_state=42)

mp = MPNeuron()
mp.entrenar(X_train, y_train)
preds = mp.predecir(X_test)

print(f'Umbral óptimo: {mp.umbral}')
print(f'Precisión: {accuracy_score(y_test, preds):.2%}')
print(f'\nReporte:\n{classification_report(y_test, preds)}')
```
