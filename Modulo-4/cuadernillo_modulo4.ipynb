## ⚡ A. Funciones de Activación

| Función | Fórmula | Rango | Derivada | Cuándo usarla |
|---|---|---|---|---|
| **ReLU** | $f(x) = \max(0, x)$ | $[0, \infty)$ | $0\text{ si }x<0,\; 1\text{ si }x>0$ | **Default** en capas ocultas. Rápida, evita vanishing gradient |
| **Leaky ReLU** | $f(x) = \max(\alpha x, x)$ | $(-\infty, \infty)$ | $\alpha \text{ si }x<0,\; 1\text{ si }x>0$ | Cuando neuronas "mueren" con ReLU (Dying ReLU) |
| **Sigmoid** | $\sigma(x) = \frac{1}{1+e^{-x}}$ | $(0, 1)$ | $\sigma(x)(1-\sigma(x))$ | **Salida binaria**. Útil en capas ocultas de RNN clásicas |
| **Tanh** | $\tanh(x) = \frac{e^x-e^{-x}}{e^x+e^{-x}}$ | $(-1, 1)$ | $1-\tanh^2(x)$ | Capas ocultas de RNN/LSTM. Centrada en 0 → mejor que sigmoid |
| **Softmax** | $\sigma(\vec{z})_i = \frac{e^{z_i}}{\sum_j e^{z_j}}$ | $(0, 1)$, suma=1 | Ver libro | **Salida multiclase**. Convierte logits en probabilidades |
| **Linear** | $f(x) = x$ | $(-\infty, \infty)$ | $1$ | **Salida de regresión** (valores continuos) |

> 💡 **Regla rápida:** ReLU en ocultas, Softmax/Sigmoid en salida de clasificación, Linear en regresión.

## 📊 D. Métricas de Evaluación

### Clasificación

| Métrica | Fórmula | Enfócate en ella cuando… |
|---|---|---|
| **Accuracy** | $(TP+TN)/(TP+TN+FP+FN)$ | Clases balanceadas |
| **Precision** | $TP/(TP+FP)$ | Minimizar **falsos positivos** (ej. spam, diagnósticos) |
| **Recall** | $TP/(TP+FN)$ | Minimizar **falsos negativos** (ej. cáncer, fraudes) |
| **F1-Score** | $2\cdot\frac{P\cdot R}{P+R}$ | Balance precision/recall en clases desbalanceadas |
| **Specificity** | $TN/(TN+FP)$ | Qué tan bien identifica los negativos |

### Regresión

| Métrica | Fórmula | Interpretación |
|---|---|---|
| **MSE** | $\frac{1}{N}\sum(y-\hat{y})^2$ | Penaliza errores grandes (cuadrático) |
| **RMSE** | $\sqrt{MSE}$ | En las mismas unidades que $y$ |
| **MAE** | $\frac{1}{N}\sum|y-\hat{y}|$ | Error promedio absoluto, más robusto |
| **R²** | $1 - \frac{SS_{res}}{SS_{tot}}$ | Proporción de varianza explicada (0 a 1) |

## 🔧 F. Regularización — Combatiendo el Overfitting

| Técnica | Código | Efecto |
|---|---|---|
| **Dropout** | `layers.Dropout(0.3)` | Desactiva neuronas aleatoriamente → red más robusta |
| **L2 (Weight Decay)** | `layers.Dense(64, kernel_regularizer='l2')` | Penaliza pesos grandes → pesos pequeños y suaves |
| **L1** | `layers.Dense(64, kernel_regularizer='l1')` | Penaliza pesos → pesos dispersos (muchos en 0) |
| **BatchNormalization** | `layers.BatchNormalization()` | Normaliza activaciones → entrenamiento más estable y rápido |
| **Early Stopping** | `EarlyStopping(patience=5)` | Detiene entrenamiento cuando val_loss deja de mejorar |
| **ReduceLROnPlateau** | `ReduceLROnPlateau(factor=0.5)` | Baja lr cuando la pérdida se estanca |
| **Data Augmentation** | `RandomFlip(), RandomRotation()` | Genera variaciones artificiales de los datos |

```python
# Ejemplo: varias técnicas juntas
modelo = keras.Sequential([
    layers.Dense(128, activation='relu', kernel_regularizer='l2'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(num_clases, activation='softmax')
])
```

### Diagnóstico rápido

| Síntoma | Problema | Solución |
|---|---|---|
| Train loss baja ✅, Val loss sube 📈 | **Overfitting** | Más dropout, L2, early stopping, más datos |
| Ambas losses altas y planas 📉 | **Underfitting** | Más capas/neuronas, más epochs, lr más alto, menos dropout |
| Loss no baja desde el inicio | **lr muy alto / bug** | Bajar lr, revisar datos (¿normalizados?) |
| Train loss > Val loss ❓ | **Fuga de datos** | Revisar train_test_split (¿data leakage?) |

| `layers.MaxPooling2D` | `MaxPooling2D(pool_size)` | `pool_size` | `MaxPooling2D((2,2))` |
| `layers.LSTM` | `LSTM(units, return_sequences)` | `units, return_sequences, bidirectional` | `LSTM(64, return_sequences=True)` |
| `layers.Embedding` | `Embedding(input_dim, output_dim)` | `input_dim, output_dim, mask_zero` | `Embedding(10000, 64)` |
| `layers.Dropout` | `Dropout(rate)` | `rate`: fracción a desactivar | `Dropout(0.3)` |
| `layers.BatchNormalization` | `BatchNormalization()` | — | `BatchNormalization()` |
| `layers.Flatten` | `Flatten()` | — | `Flatten()` |
| `layers.GlobalAveragePooling2D` | `GlobalAveragePooling2D()` | — | `GlobalAveragePooling2D()` |

### Compilación, entrenamiento, evaluación

| Método | Firma simplificada | Parámetros clave | Ejemplo |
|---|---|---|---|
| `model.compile()` | `compile(optimizer, loss, metrics)` | `optimizer, loss, metrics` | `compile('adam', 'mse', ['mae'])` |
| `model.fit()` | `fit(x, y, epochs, batch_size, validation_data, callbacks)` | `validation_data, callbacks, verbose` | `fit(X_train, y_train, epochs=50)` |
| `model.predict()` | `predict(x)` | `x`: datos de entrada | `predict(X_test)` → probabilidades |
| `model.evaluate()` | `evaluate(x, y)` | `x, y, verbose` | `evaluate(X_test, y_test)` → `[loss, acc]` |
| `model.summary()` | `summary()` | — | `summary()` → arquitectura + parámetros |

## 🐛 I. Diagnóstico de Errores Comunes

| Error | Causa probable | Solución |
|---|---|---|
| `Input 0 of layer "dense" is incompatible` | Shape de entrada incorrecta | Revisar `input_shape` del modelo vs forma de X |
| `Failed to find data adapter` | Tipo de dato no soportado | Convertir X, y a numpy arrays con `np.array()` |
| `nan en loss` | Overflow numérico, lr muy alto | Bajar lr, revisar datos (¿infinitos?), añadir clamp |
| `GPU out of memory` | Batch size muy grande | Reducir `batch_size` (16 o 8), usar menos filtros |
| `ModuleNotFoundError: No module named 'tensorflow'` | TF no instalado | `pip install tensorflow` |
| Training muy lento | Sin GPU, batch grande | Verificar GPU, reducir batch_size, usar `.flow_from_directory` |
| Accuracy no mejora de aleatorio | Error en datos (labels) | Revisar correspondencia X/y, normalizar inputs |
| `Found 0 images` en generator | Ruta incorrecta | Verificar que las subcarpetas tengan los nombres de clase |
| Overfitting severo | Muy pocos datos, modelo muy grande | Data augmentation, más dropout, transfer learning |
| Colab se desconecta | Sesión inactiva >90min | Guardar modelos frecuentemente con `guardar_modelo()` |
