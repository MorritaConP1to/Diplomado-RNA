## 🧠 A. Fundamentos de Redes Neuronales

| Término | Definición |
|---|---|
| **Neurona artificial** | Unidad de cómputo básica: recibe entradas, aplica pesos, suma y pasa por una función de activación. Inspirada en la neurona biológica pero altamente simplificada. |
| **Capa (Layer)** | Conjunto de neuronas que procesan datos en paralelo. Una red tiene: capa de entrada, capas ocultas y capa de salida. |
| **Perceptrón** | La red neuronal más simple: una sola neurona con entradas, pesos y un umbral. Solo puede separar clases linealmente. |
| **MLP (Perceptrón Multicapa)** | Red con múltiples capas densas. Puede aprender relaciones no lineales gracias a las funciones de activación. |
| **Capa densa (Dense / FC)** | Fully Connected — cada neurona está conectada a todas las de la capa anterior. La capa más básica en Keras. |
| **Peso (Weight, W)** | Parámetro aprendible que escala cada conexión entre neuronas. La red aprende ajustando estos valores. |
| **Sesgo (Bias, b)** | Parámetro adicional por neurona que permite desplazar la activación independientemente de las entradas. |
| **Parámetro** | Todo valor que el modelo aprende durante el entrenamiento (pesos + sesgos). `model.summary()` los muestra. |

---
## ⚡ B. Funciones de Activación

| Función | Fórmula | Cuándo usarla |
|---|---|---|
| **ReLU** | `max(0, x)` | Capas ocultas — default casi siempre. Rápida, evita vanishing gradient. |
| **Leaky ReLU** | `x si x>0, αx si x≤0` | Cuando muchas neuronas "mueren" con ReLU (se quedan en 0). |
| **Sigmoid** | `1 / (1 + e^-x)` → rango (0,1) | Capa de salida de **clasificación binaria**. |
| **Softmax** | Convierte logits en probabilidades que suman 1 | Capa de salida de **clasificación multiclase**. |
| **Tanh** | `(e^x - e^-x)/(e^x + e^-x)` → rango (-1,1) | Capas ocultas de RNNs. Centrada en 0. |
| **Linear (identidad)** | `f(x) = x` | Capa de salida de **regresión** (predice valores continuos). |

> 💡 **Regla rápida:** ReLU en capas ocultas, Softmax/Sigmoid en salida de clasificación, Linear en salida de regresión.

## 📉 C. Entrenamiento: Backpropagation y Optimización

| Término | Definición |
|---|---|
| **Forward pass** | Datos fluyen de entrada → salida para generar una predicción ŷ. |
| **Loss (función de pérdida)** | Mide qué tan equivocado está el modelo. El objetivo es minimizarla. |
| **Backward pass** | Backpropagation: calcula el gradiente de la pérdida con respecto a cada peso usando la regla de la cadena. |
| **Gradiente** | Derivada parcial ∂Loss/∂W — indica en qué dirección y magnitud mover cada peso para reducir la pérdida. |
| **Descenso por gradiente (GD)** | Actualización de pesos: `W = W - lr × gradiente`. Mueve pesos en dirección contraria al gradiente. |
| **SGD (Stochastic GD)** | Variante que usa un solo ejemplo (o un mini-batch) por actualización. Más ruidoso pero más rápido. |
| **Adam** | Optimizador adaptativo (combina momentum + RMSProp). Default recomendado para la mayoría de redes. lr=0.001. |
| **Learning rate (lr)** | Tamaño del paso de actualización. Muy alto → no converge. Muy bajo → entrena muy lento. |
| **Epoch** | Una pasada completa sobre todo el dataset de entrenamiento. |
| **Batch size** | Número de muestras procesadas antes de actualizar los pesos. Típico: 32 o 64. |
| **Iteración** | Una actualización de pesos (= 1 batch procesado). Epochs × (n_muestras/batch_size) = total iteraciones. |
| **Vanishing gradient** | El gradiente se hace tan pequeño al retropropagar que las capas iniciales casi no aprenden. Problema en redes profundas y RNNs. |
| **Exploding gradient** | El gradiente crece demasiado. Se controla con gradient clipping. |

## 🎛️ D. Hiperparámetros

| Hiperparámetro | Qué controla | Valores típicos |
|---|---|---|
| `learning_rate` | Velocidad de aprendizaje | 0.1 (SGD), 0.001 (Adam) |
| `batch_size` | Muestras por actualización | 32, 64, 128 |
| `epochs` | Rondas de entrenamiento | 10–100 (con early stopping) |
| `n_layers` | Profundidad de la red | 2–5 capas (más no siempre es mejor) |
| `n_units` | Neuronas por capa | 32, 64, 128, 256 |
| `dropout` | % neuronas a desactivar aleatoriamente | 0.2–0.5 |
| `activation` | Función de activación por capa | relu, tanh, sigmoid |
| `optimizer` | Algoritmo de actualización | Adam, SGD, RMSProp |
| `kernel_size` | Tamaño del filtro en Conv2D | (3,3), (5,5) |
| `filters` | Número de filtros en Conv2D | 32, 64, 128 |

## 🔧 E. Regularización y Control de Overfitting

| Técnica | Cómo funciona | En Keras |
|---|---|---|
| **Dropout** | Desactiva neuronas aleatoriamente en cada batch. Fuerza a la red a no depender de neuronas específicas. | `layers.Dropout(0.3)` |
| **L2 (Weight Decay)** | Penaliza pesos grandes en la función de pérdida. | `layers.Dense(64, kernel_regularizer='l2')` |
| **BatchNormalization** | Normaliza las activaciones de cada capa. Estabiliza y acelera el entrenamiento. | `layers.BatchNormalization()` |
| **Early Stopping** | Detiene el entrenamiento cuando val_loss deja de mejorar. | `callbacks.EarlyStopping(patience=5)` |
| **Data Augmentation** | Genera variaciones artificiales de los datos de entrenamiento. Solo imágenes/texto. | `layers.RandomFlip()`, etc. |
| **Reducir lr** | Baja el learning rate cuando la pérdida se estanca. | `callbacks.ReduceLROnPlateau()` |

### Diagnóstico:
| Síntoma | Problema | Solución |
|---|---|---|
| Train loss baja, Val loss sube | **Overfitting** | Más dropout, L2, menos epochs, más datos |
| Ambas losses altas y planas | **Underfitting** | Más capas, más neuronas, más epochs, lr más alto |
| Loss no baja desde el inicio | **lr muy alto / bug** | Bajar lr, revisar datos |

## 🖼️ F. Redes Convolucionales (CNNs)

| Término | Definición |
|---|---|
| **Convolución** | Operación matemática que aplica un filtro pequeño sobre la imagen para detectar patrones locales. |
| **Filtro (Kernel)** | Matriz pequeña de pesos (ej. 3×3) que se desplaza sobre la imagen. Cada filtro detecta un patrón diferente. |
| **Feature map** | Resultado de aplicar un filtro a una imagen. Muestra dónde se detectó el patrón. |
| **Stride** | Cuántos píxeles se desplaza el filtro en cada paso. Mayor stride = feature map más pequeño. |
| **Padding** | Relleno de ceros alrededor de la imagen. `same` conserva dimensiones, `valid` las reduce. |
| **Pooling** | Reduce el tamaño espacial tomando el máximo (`MaxPooling`) o promedio (`AveragePooling`) de una región. |
| **Flatten** | Convierte el tensor 3D (alto×ancho×canales) en un vector 1D para conectar con capas Dense. |
| **GlobalAveragePooling** | Alternativa a Flatten: promedia cada feature map en un solo valor. Menos parámetros. |
| **Transfer Learning** | Reusar pesos de una CNN entrenada en ImageNet para una nueva tarea. Se congela la base y se entrena solo la cabeza. |
| **Fine-tuning** | Descongelar capas de la base y re-entrenar con lr muy bajo. Segunda fase del transfer learning. |
| **ImageNet** | Dataset de 1.2M imágenes y 1000 clases. Benchmark estándar para CNNs. |
| **VGG16/19** | CNN profunda (16/19 capas). Arquitectura clásica, simple pero pesada. |
| **ResNet** | Redes Residuales: introducen conexiones de salto (skip connections) que resuelven el vanishing gradient en redes muy profundas. |
| **MobileNet** | CNN ligera optimizada para dispositivos móviles. Usa convoluciones depthwise separables. |

## 📝 G. Redes Recurrentes (RNNs / LSTMs / GRU)

| Término | Definición |
|---|---|
| **RNN (Red Recurrente)** | Red con memoria: el estado oculto del paso anterior se pasa al siguiente. Procesa secuencias. |
| **Estado oculto (h)** | Vector que resume lo que la RNN recuerda de los pasos anteriores. Se actualiza en cada paso. |
| **LSTM** | Long Short-Term Memory. Resuelve el vanishing gradient con puertas que controlan qué recordar/olvidar. |
| **GRU** | Gated Recurrent Unit. Versión simplificada de LSTM, más rápida y casi igual de efectiva. |
| **Puerta (Gate)** | Mecanismo en LSTM/GRU que controla el flujo de información. Puertas: forget, input, output (LSTM). |
| **Celda de estado (C)** | Vector de memoria a largo plazo en LSTM. Se modifica solo con operaciones suaves (suma/multiplicación). |
| **Bidireccional** | Procesa la secuencia en ambas direcciones (izquierda→derecha y derecha→izquierda) para más contexto. |
| **return_sequences** | Si es True, la capa LSTM devuelve la secuencia completa de estados ocultos (no solo el último). |
| **Embedding** | Capa que convierte índices de palabras en vectores densos de tamaño fijo. Aprende representaciones semánticas. |
| **Padding** | En texto: rellenar secuencias con 0s para que todas tengan la misma longitud. |
| **Tokenización** | Convertir texto en secuencias de enteros (índices de palabras en el vocabulario). |
| **OOV (Out of Vocabulary)** | Palabras que no estaban en el vocabulario de entrenamiento. Se manejan con un token especial. |

## 🌐 H. Procesamiento de Lenguaje Natural (PLN / NLP)

| Término | Definición |
|---|---|
| **Corpus** | Colección de textos usada para entrenar o evaluar un modelo de PLN. |
| **Vocabulario** | Conjunto de todas las palabras únicas en el corpus. |
| **Word Embedding** | Representación vectorial densa de palabras que captura relaciones semánticas. ej. rey - hombre + mujer ≈ reina. |
| **Word2Vec** | Algoritmo que aprende embeddings de palabras a partir de contexto. Produce vectores con significado geométrico. |
| **GloVe** | Global Vectors — embeddings preentrenados en corpus grandes. Se pueden cargar sin entrenar desde cero. |
| **Sentiment Analysis** | Tarea de clasificar el sentimiento de un texto (positivo/negativo/neutro). |
| **Attention** | Mecanismo que permite al modelo enfocarse en partes específicas de la secuencia de entrada. Base de Transformers. |
| **Transformer** | Arquitectura basada en attention (sin recurrencia). Base de BERT, GPT, y modelos modernos de PLN. |
| **BERT** | Bidirectional Encoder Representations from Transformers. Modelo preentrenado de Google para tareas NLP. |
| **Tokenización subpalabra** | Divide palabras en fragmentos (ej. "unboxing" → ["un", "##box", "##ing"]) para manejar OOV mejor. |

## 📊 I. Métricas de Evaluación

| Métrica | Fórmula simplificada | Cuándo enfocarse en ella |
|---|---|---|
| **Accuracy** | (TP+TN) / total | Clases balanceadas |
| **Precision** | TP / (TP+FP) | Minimizar falsos positivos (ej. spam) |
| **Recall** | TP / (TP+FN) | Minimizar falsos negativos (ej. diagnóstico médico) |
| **F1-Score** | 2×(P×R)/(P+R) | Balance precision/recall. Clases desbalanceadas. |
| **Loss** | Depende del problema | Siempre monitorear. Separada de accuracy. |
| **MAE** | Mean Absolute Error | Regresión, unidades interpretables |
| **MSE/RMSE** | Mean Squared Error | Regresión, penaliza errores grandes |

### Funciones de pérdida por tipo de tarea:
| Tarea | Función de pérdida | Activación de salida |
|---|---|---|
| Clasificación binaria | `binary_crossentropy` | Sigmoid |
| Clasificación multiclase (etiquetas enteras) | `sparse_categorical_crossentropy` | Softmax |
| Clasificación multiclase (etiquetas OHE) | `categorical_crossentropy` | Softmax |
| Regresión | `mse` o `mae` | Linear |

## 🔤 K. Términos de Keras / TensorFlow

| Término | Definición |
|---|---|
| **`keras.Sequential`** | Forma más simple de construir modelos: capas apiladas una tras otra. |
| **`keras.Model` (Functional API)** | Permite crear modelos con bifurcaciones, múltiples entradas/salidas. Más flexible que Sequential. |
| **`model.compile()`** | Configura optimizador, función de pérdida y métricas antes de entrenar. |
| **`model.fit()`** | Entrena el modelo. Devuelve un objeto `history` con las métricas por epoch. |
| **`model.predict()`** | Genera predicciones. Devuelve probabilidades (no clases). |
| **`model.evaluate()`** | Calcula métricas sobre un conjunto de datos (sin actualizar pesos). |
| **`model.summary()`** | Muestra arquitectura, capas y número de parámetros. |
| **Callback** | Función que se ejecuta en puntos específicos del entrenamiento (inicio/fin de epoch, batch, etc.). |
| **`EarlyStopping`** | Callback que detiene el entrenamiento si val_loss no mejora. |
| **`ModelCheckpoint`** | Callback que guarda el mejor modelo durante el entrenamiento. |
| **Tensor** | Array multidimensional. Unidad de datos en TensorFlow. |
| **`tf.data.Dataset`** | Pipeline eficiente para cargar y preprocesar datos en TF. Mejor para datasets grandes. |
