## 🚀 Despliegue de Modelos

Pasos para llevar un modelo a producción:

1. **Entrenar y validar** el modelo
2. **Guardar** el modelo (formato .keras, SavedModel)
3. **Convertir** a formato de inferencia (TFLite, ONNX, TensorRT)
4. **Servir** el modelo (API REST, endpoint, app móvil)
5. **Monitorear** drift, latencia, precisión

---
## 🔄 MLOps Concepts

- **Versionado de modelos:** Trackear cada versión con metadatos
- **Pipeline CI/CD:** Automatizar entrenamiento, prueba y despliegue
- **Feature Store:** Repositorio centralizado de features
- **Model Registry:** Catálogo de modelos con métricas y estado
- **Drift Detection:** Monitorear cambios en datos y predicciones

````python
# MACHOTE: Configuración inicial
import sys, os
sys.path.insert(0, os.path.abspath(os.path.join('..')))
from modulo4_libreria import *

INFO = setup_completo()
````

````python
# MACHOTE: setup_completo() ya detecta Colab vs Local automáticamente
# La función INFO contiene toda la información del entorno

print(f"¿Estamos en Google Colab?  {INFO['en_colab']}")
print(f"Plataforma detectada:      {INFO['plataforma']}")
print(f"Semilla de reproducibilidad: {INFO['semilla']}")

if INFO['gpu']:
    print(f"GPU(s) detectadas: {len(INFO['gpu'])}")
else:
    print("⚠️  No se detectó GPU")

print("\nRutas configuradas:")
for k, v in INFO['rutas'].items():
    print(f"  {k}: {v}")
````

````python
# MACHOTE: Guardar un modelo entrenado y verificar archivo
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=500, n_features=10, n_classes=2, random_state=42)
datos = train_val_test_split(X, y, test_size=0.2, val_size=0.1)

modelo_export = crear_mlp((10,), 2, capas_ocultas=[64], dropout=0.2)
hist = compilar_y_entrenar(modelo_export, datos['X_train'], datos['y_train'],
                           datos['X_val'], datos['y_val'],
                           num_clases=2, lr=0.001, epochs=10, batch_size=32,
                           verbose=0, early_stop_paciencia=10)

ruta_modelo = guardar_modelo(modelo_export, nombre='modelo_exportado',
                             ruta_modelos=INFO['rutas']['modelos'])

# Verificar que el archivo existe
if os.path.exists(ruta_modelo):
    tamano = os.path.getsize(ruta_modelo)
    print(f"\n✅ Archivo verificado: {ruta_modelo}")
    print(f"📦 Tamaño: {tamano / 1024:.2f} KB")
````
