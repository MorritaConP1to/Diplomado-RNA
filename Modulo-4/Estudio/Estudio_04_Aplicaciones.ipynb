## 🔄 Transfer Learning

Consiste en tomar un modelo preentrenado (ej. en ImageNet) y reutilizar sus pesos para una tarea nueva. Ventajas:
- Requiere menos datos
- Entrena más rápido
- Mejor precisión que desde cero

```python
# MACHOTE: Configuración inicial
import sys, os
sys.path.insert(0, os.path.abspath(os.path.join('..')))
from modulo4_libreria import *

INFO = setup_completo()
```

```python
# MACHOTE: Transfer Learning con MobileNetV2 (5 clases)
modelo_tl, base_tl = crear_transfer_learning(
    num_clases=5,
    input_shape=(224, 224, 3),
    base='MobileNetV2',
    trainable_base=False
)

print("\n=== MODELO COMPLETO (Transfer Learning) ===")
modelo_tl.summary()

print("\n=== BASE (MobileNetV2 - congelada) ===")
print(f"Capas en base: {len(base_tl.layers)}")
print(f"Base entrenable: {base_tl.trainable}")
```

```python
# MACHOTE: Data Augmentation — crear y visualizar
augmentation = crear_data_augmentation()

print("Pipeline de aumento de datos:")
for capa in augmentation.layers:
    print(f"  - {capa.name}")

# Cargar MNIST para visualizar aumentos
(X_muestra, _), _ = keras.datasets.mnist.load_data()
imagen_original = X_muestra[0].reshape(1, 28, 28, 1).astype('float32') / 255.0

# Redimensionar a 224x224x3 para que funcione con la augmentation
imagen_grande = tf.image.resize(imagen_original, (224, 224))
imagen_rgb = tf.image.grayscale_to_rgb(imagen_grande)

# Aplicar aumentos
plt.figure(figsize=(12, 4))
plt.subplot(1, 5, 1)
plt.imshow(imagen_rgb[0].numpy().squeeze(), cmap='gray')
plt.title('Original'); plt.axis('off')

for i in range(4):
    img_aug = augmentation(imagen_rgb)
    plt.subplot(1, 5, i+2)
    plt.imshow(img_aug[0].numpy().squeeze(), cmap='gray')
    plt.title(f'Augment {i+1}'); plt.axis('off')

plt.suptitle('Data Augmentation - Ejemplo visual', fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()
```

```python
# MACHOTE: Preprocesamiento de texto con nuestra librería
oraciones = [
    "Me encantó esta película, es increíblemente buena",
    "La comida estuvo horrible, no vuelvo nunca",
    "El servicio fue excelente, muy recomendable",
    "Este producto es malo y de baja calidad",
    "Una experiencia maravillosa, superó mis expectativas"
]

resultado_texto = preprocesar_texto(
    textos_train=oraciones,
    vocab_size=100,
    max_len=10
)

print(f"\nDiccionario (primeras 10 palabras):")
palabras = list(resultado_texto['tokenizador'].word_index.items())[:10]
for palabra, idx in palabras:
    print(f"  {idx}: {palabra}")

print(f"\nSecuencia padded de 'Me encantó esta película...':")
print(resultado_texto['X_train_pad'][0])
```
