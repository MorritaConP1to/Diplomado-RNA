## 🧠 Tema 1: Fundamentos de Redes Neuronales

### MLP (Perceptrón Multicapa) en 30 segundos
1. Entrada → Capas densas con ReLU → Dropout → Salida Softmax  
2. `input_shape = (n_features,)`  
3. `num_clases = 2` → Sigmoid, `>2` → Softmax

````python
# Ejemplo rápido: MLP con datos sintéticos
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=20, n_classes=3, random_state=42)
data = train_val_test_split(X, y)

modelo = crear_mlp(input_shape=(20,), num_clases=3, capas_ocultas=[64, 32])
modelo.summary()

hist = compilar_y_entrenar(modelo, data['X_train'], data['y_train'],
                            data['X_val'], data['y_val'], num_clases=3, epochs=20)

graficar_historia(hist, "MLP — Clasificación")
````

---
## 🖼️ Tema 2: CNN — Visión por Computadora

### CNN en 30 segundos
1. Conv2D(32, (3,3)) + BatchNorm + MaxPooling(2,2) — detecta bordes  
2. Conv2D(64, (3,3)) + BatchNorm + MaxPooling(2,2) — detecta formas  
3. GlobalAveragePooling2D → Dense(128) → Dropout → Salida  
4. `input_shape = (alto, ancho, canales)`

````python
# Ejemplo rápido: CNN con MNIST
(X_train, y_train), (X_test_raw, y_test_raw) = keras.datasets.mnist.load_data()

X_train = X_train.astype('float32') / 255.0
X_test  = X_test_raw.astype('float32') / 255.0
X_train = X_train[..., np.newaxis]
X_test  = X_test[..., np.newaxis]

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

cnn = crear_cnn(input_shape=(28, 28, 1), num_clases=10)
cnn.summary()

hist = compilar_y_entrenar(cnn, X_train, y_train, X_val, y_val, num_clases=10, epochs=10)
graficar_historia(hist, "CNN — MNIST")
````

---
## 📝 Tema 3: RNN / LSTM — PLN y Series de Tiempo

### LSTM en 30 segundos
1. Embedding(vocab_size, dim) → convierte índices a vectores densos  
2. Bidirectional(LSTM(64)) → captura contexto izq→der y der→izq  
3. Dense + Dropout → Salida  
4. `input_shape = (max_len,)` — enteros (índices de palabras)

````python
# Ejemplo rápido: LSTM con IMDB
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocab_size = 10000
max_len = 200

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)
X_train = pad_sequences(X_train, maxlen=max_len)
X_test  = pad_sequences(X_test, maxlen=max_len)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

lstm = crear_lstm_texto(vocab_size, max_len, embedding_dim=64, num_clases=2)
lstm.summary()

hist = compilar_y_entrenar(lstm, X_train, y_train, X_val, y_val, num_clases=2, epochs=5)
graficar_historia(hist, "LSTM — IMDB Sentiment")
````
